# Merge Operations

```python
data4 = {
    'emp_id' : [101 , 102  , 103 ,104],
    'salary' : [234 , 345 , 200 , 400],
    'pf' : [23,34,56,78]
}
```

```python
data5 = {
    'emp_id' : [101 , 102 , 104 ,105],
    'mob_no' : [3434 , 5656 ,767 , 878],
    'house_nu' : [345 , 345 , 345, 56]
}
```

```python
df4 = pd.DataFrame(data4)
df5 = pd.DataFrame(data5)
```

```python
df4
```

```python
df5
```

```python
pd.merge(df4,df5)
```

```python
pd.merge(df4,df5 , how = 'left')
```

```python
pd.merge(df4,df5 , how = 'right')
```

```python
pd.merge(df4,df5 , how = 'right' , on = 'emp_id')
```

```python
data6 = {
    'emp_id1' : [101 , 102  , 103 ,104],
    'salary' : [234 , 345 , 200 , 400],
    'pf' : [23,34,56,78]
}
data7 = {
    'emp_id2' : [101 , 102 , 104 ,105],
    'mob_no' : [3434 , 5656 ,767 , 878],
    'house_nu' : [345 , 345 , 345, 56]
}
```

```python
df6 = pd.DataFrame(data6)
df7 = pd.DataFrame(data7)
```

```python
pd.merge(df6,df7)
```

```python
pd.merge(df6,df7 , left_on = 'emp_id1' , right_on = 'emp_id2' , how = 'inner')
```

# Load Dataset

```python
df_sales = pd.read_csv('C:/Users/91942/Data Science/Datasets/sales_data_final.csv')
```

```python
def profit_flag(a):
    if a>0:
        return 'positive'
    else:
        return 'negative'
```

```python
profit_flag(-45)
```

```python
profit_flag(45)
```

```python
df_sales['flag_profit'] = df_sales['profit'].apply(profit_flag)
```

```python
df_sales.head()
```

```python
df_sales['len_cust_name'] = df_sales['customer_name'].apply(len)
```

```python
df_sales.head()
```

```python
def quantity_flag(a):
    if a<10:
        return 'low'
    elif a>10 and a<20:
        return 'medium'
    else:
        return 'high'
```

```python
df_sales['flag_quantiy'] = df_sales['quantity'].apply(quantity_flag)
```

```python
df_sales.head(20)
```

```python
df_sales['square_quantity'] = df_sales['quantity'].apply(lambda a: a**2)
```
