# 3D plot

```python
from mpl_toolkits import mplot3d
```

```python
x = np.linspace(-1 , 6 , 30)
x
```

```python
y = np.linspace(-1,6 , 30)
y
```

```python
z = x + y
z
```

```python
ax = plt.axes(projection = '3d')
ax.plot3D(x , y, z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
ax = plt.axes(projection = '3d')
ax.plot3D(df['sepal_length'] , df['sepal_width'], df['petal_length'])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
